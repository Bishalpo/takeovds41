```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def convert_duration_to_minutes(duration_str):
    if pd.isna(duration_str):
        return None
    parts = duration_str.split(' ')
    hours = int(parts[0].replace('h', '')) if 'h' in parts[0] else 0
    minutes = int(parts[1].replace('m', '')) if len(parts) > 1 else 0
    return hours * 60 + minutes

Imbd_data['duration_minutes'] = Imbd_data['duration'].apply(convert_duration_to_minutes)

# Calculate the average duration
average_duration = Imbd_data['duration_minutes'].mean()
average_duration
```

```python
Imbd_data['duration'].fillna(average_duration, inplace=True)

Imbd_data.head()
```

```python
missing_duration = Imbd_data[Imbd_data['duration'].isna()]
print(missing_duration)
```

```python
# Checking the trend of average movie ratings over the years for the dataset of lowest-ranked movies:

Imbd_data['year'] = pd.to_numeric(Imbd_data['year'], errors='coerce')

yearly_data = Imbd_data.groupby('year')['rating'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x='year', y='rating', data=yearly_data)
plt.title('Average Movie Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
# Analyzing the distribution of movie ratings

Imbd_data['rating'] = pd.to_numeric(Imbd_data['rating'], errors='coerce')

plt.figure(figsize=(10, 5))
sns.histplot(Imbd_data['rating'], bins=30, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Checking if there's a correlation between the number of reviews and movie ratings

# Removing 'K' and multiply the number by 1000.
Imbd_data['review_count'] = Imbd_data['review_count'].str.replace('K', '').astype(float) * 1000

# Calculating the correlation between number of reviews and movie ratings
correlation = Imbd_data[['review_count', 'rating']].corr()

# Plotting the relationship between number of reviews and movie ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_count', y='rating', data=Imbd_data)
plt.title('Relationship Between Number of Reviews and Movie Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.grid(True)
plt.show()

correlation
```

```python
from collections import Counter
import ast

# Flatten the list of genres
flat_genres = [genre for sublist in Imbd_data['genre'] for genre in sublist]

# Counting the frequency of each genre
genre_counts = Counter(flat_genres)

# Converting the Counter object to a DataFrame for better visualization
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index', columns=['Count']).sort_values(by='Count', ascending=False)

# Plotting the frequency of genres in low-rated movies
plt.figure(figsize=(10, 5))
sns.barplot(x=genre_df['Count'], y=genre_df.index)
plt.title('Frequency of Genres in Low-Rated Movies')
plt.xlabel('Frequency')
plt.ylabel('Genre')
plt.show()
```
