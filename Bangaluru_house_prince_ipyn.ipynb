```python

import re

def convert_range_to_average(value):

    if '-' in str(value):
        range_values = re.findall(r'\d+\.?\d*', value)
        if len(range_values) == 2:
            average_value = sum(map(float, range_values)) / 2
            return average_value
    return value

house_data_filled['total_sqft'] = house_data_filled['total_sqft'].apply(convert_range_to_average)

house_data_filled
```

```python
missing_values = house_data_filled.isnull().sum()
missing_values
```

```python
def is_range(value):
    if re.match(r'\b\d+\-\d+\b', str(value)):
        return True
    return False

ranges_in_total_sqft = house_data_filled[house_data_filled['total_sqft'].apply(is_range)]

print(ranges_in_total_sqft)
```

```python

import re
def convert_to_sqft(total_sqft):
    # Dictionary for conversion factors
    conversion_factors = {
        'acres': 43560,
        'sq. meter': 10.7639,
        'sq. yard': 9,
        'cents': 435.6,
        'grounds': 2400,
        'guntha': 1089,
        'perch': 272.25
    }

    # Regular expression to find numbers and units
    pattern = re.compile(r'(\d+\.?\d*)\s*([\w\s.]+)')
    match = pattern.match(str(total_sqft).lower())

    if match:
        number, unit = match.groups()
        number = float(number)

        # Convert to square feet if a matching unit is found
        if unit in conversion_factors:
            return number * conversion_factors[unit]
        else:
            return number  # Assume the unit is already in square feet
    else:
        return None
house_data_filled['total_sqft'] = house_data_filled['total_sqft'].apply(convert_to_sqft)
house_data_filled
```

```python
missing_values = house_data_filled.isnull().sum()
missing_values
```

```python
missing_sqft_rows = house_data_filled[house_data_filled['total_sqft'].isna()]
print(missing_sqft_rows)
```

```python
mean_total_sqft = house_data_filled['total_sqft'].mean()

# Replace NaN values in the 'total_sqft' column with the calculated mean
house_data_filled['total_sqft'].fillna(mean_total_sqft, inplace=True)

house_data_filled
```
